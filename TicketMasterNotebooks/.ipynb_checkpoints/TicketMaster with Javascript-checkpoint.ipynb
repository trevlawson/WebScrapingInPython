```python
events = driver.find_elements_by_xpath("//li[starts-with(@class, 'event')]")
len(events)
```

```python
def extract_event(event, columns):
    source = event.get_attribute('innerHTML')
    lil_soup = BeautifulSoup(source, 'lxml')
    
    name = lil_soup.find(
            "div", 
            attrs={'data-aqa-id': lambda value: value and value.startswith("event-name")}
    ).text

    date_time = lil_soup.find(
            "div", 
            attrs={'data-aqa-id': lambda value: value and value.startswith("event-date")}
        )
    date, time = map(lambda element: element.text, date_time.findChildren())

    venue_info = lil_soup.find(
            "div", 
            attrs={'data-aqa-id': lambda value: value and value.startswith("event-venue")}
        )
    venue, city_state = map(lambda element: element.text, venue_info.findChildren())

    return pd.Series([name, date, time, venue, city_state], index=columns)
```

```python
import pandas as pd

columns = ['Name', 'Date', 'Time', 'Venue', 'City_State']
df = pd.DataFrame(columns=columns)

for event in events[1:]:
    try:
        s = extract_event(event, columns)
        df = df.append(s, ignore_index=True)
    except:
        print(event.text)

df[['Date', 'Time', 'Name', 'Venue', 'City_State']]
```
